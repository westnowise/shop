```python
# https://datarian.io/blog/what-is-rfm
# 1. 입점 기업별 RFM 분석(3그룹) 진행
# 얼마나 최근까지 판매했는가
# 얼마나 자주 구매했는가
# 얼마나 많은 금액을 지출했는가
# 분포를 확인하기
# 가설 : 자주 상위 5개에 랭크되어 있는 업체 3곳이 가장 RFM 점수가 높을것이다 

# 주문일자를 날짜 형식으로 변환
shop_df['주문일자'] = pd.to_datetime(shop_df['주문일자'])# 날짜 형식으로 변환

# 중복 제거 후 최근 판매 나열
recently_sold = shop_df.sort_values(by='주문일자', ascending=False).drop_duplicates('업체명')[['업체명', '주문일자']]

# 현재 날짜 계산
current_date = pd.to_datetime('2022-11-09')
recently_sold['Recency'] = (current_date - recently_sold['주문일자']).dt.days

# 등급 부여
recently_sold['RecencyGrade'] = pd.cut(recently_sold['Recency'], bins=[0, 90, 180, 365, 730, float('inf')],
                                       labels=[5, 4, 3, 2, 1], right=False)

# 결과 출력
recently_sold
```

```python
# 중복 제거 후 자주 구매 나열
frequency_sold = shop_df['업체명'].value_counts()
frequency_result = pd.DataFrame({'업체명': frequency_sold.index, 'Frequency': frequency_sold.values})

# 등급 부여 함수 정의
def assign_frequency_grade(freq_series):
    freq_series = freq_series.copy()  # 복사본 생성
    
    # 1등급: 빈도수가 1인 업체
    freq_series.loc[freq_series == 1] = 1
    
    # 2등급 이상: 1등급을 제외한 나머지 업체들에 대해 등급 지정
    for grade in range(2, 6):
        # 이전 등급을 제외한 나머지 업체들 중에서 평균을 구하기
        avg_frequency = freq_series.loc[freq_series > (grade - 1)].mean()
        # 이전 등급을 제외하고 남은 빈도수의 업체들 중에서 평균보다 낮은 그룹에게 등급 부여
        freq_series.loc[(freq_series > (grade - 1)) & (freq_series <= avg_frequency)] = grade
    
    # 평균보다 큰 나머지 빈도수의 업체에 대해 등급을 5로 지정
    freq_series.loc[freq_series > avg_frequency] = 5
    
    return freq_series

# 빈도수에 따라 등급 부여
frequency_result['FrequencyGrade'] = assign_frequency_grade(frequency_result['Frequency'])

frequency_result
```

```python
# 업체별 총 판매금액 계산 및 정렬
monetary_sold = shop_df.groupby('업체명')['판매금액'].sum().sort_values(ascending=False)

# 결과 DataFrame 생성
monetary_result = pd.DataFrame({'업체명': monetary_sold.index, 'Monetary': monetary_sold.values})

# 각 업체에서 가장 많은 판매 금액을 기록한 상품명 가져오기
most_sold_products = shop_df.loc[shop_df.groupby('업체명')['판매금액'].idxmax()][['업체명']]

# 총 판매 금액을 20% 단위로 나누어 5등급으로 매기기
result_with_products['MonetaryGrade'] = pd.qcut(result_with_products['Monetary'], q=[0, 0.2, 0.4, 0.6, 0.8, 1], labels=False, duplicates='drop') + 1

result_with_products
```

```python
# 결과 데이터프레임 합치기
merged_result = pd.merge(recently_sold, result_with_products, on='업체명')
merged_result = pd.merge(merged_result, frequency_result, on='업체명')

# RecencyGrade, MonetaryGrade, FrequencyGrade 합치기
merged_result['Grade'] = merged_result['RecencyGrade'].astype(int) + merged_result['MonetaryGrade'].astype(int) + merged_result['FrequencyGrade'].astype(int)

# 결과 출력
merged_result
```

```python
# 구매확정, 교환완료, 상담형상품접수, 환불요청, 배송완료, 배송준비, 배송중, 시안확인요청, 결제완료 만 주문금액 인정으로 처리,나머지는 주문금액 없다고 처리
# 주문취소, 미결제, 후불반려, 환불완료, 환불승인, 부분환불, 부분취소, 상담취소

# 할부기간이 있는 경우에 판매 금액과 환불금액을 나누어 주문일자로부터 금액 처리

shop_df['주문일자'] = pd.to_datetime(shop_df['주문일자'])

# 주문금액 처리
shop_df['처리된 판매금액'] = 0 

for idx, row in shop_df.iterrows():
    if row['처리상태'] in ['구매확정', '배송완료', '배송준비', '배송중', '결제완료']:
        try:
            if pd.notna(row['할부기간']):
                installment_period = float(row['할부기간'].replace('개월', ''))
                if installment_period > 0:
                    shop_df.at[idx, '처리된 판매금액'] = row['판매금액'] / installment_period
                else:
                    shop_df.at[idx, '처리된 판매금액'] = row['판매금액']   
            else:
                shop_df.at[idx, '처리된 판매금액'] = row['판매금액']
        except (ValueError, AttributeError):
            # '할부기간'이 숫자로 변환되지 않거나 NaN일 경우 처리
            shop_df.at[idx, '처리된 판매금액'] = row['판매금액']
# 판매금액을 만원 단위로 변경
shop_df['처리된 판매금액'] = shop_df['처리된 판매금액'] / 10000

# 월별 순수익 계산
shop_df['연도'] = shop_df['주문일자'].dt.year
shop_df['월'] = shop_df['주문일자'].dt.month
monthly_profit = shop_df.groupby(['연도', '월'])['처리된 판매금액'].sum()

# 그래프 그리기
fig, ax = plt.subplots(figsize=(10, 6))
for year in monthly_profit.index.levels[0]:
    plt.plot(monthly_profit.loc[year].index, monthly_profit.loc[year], label=f'{year}년', marker='o')
# 범례 추가
plt.legend(labels=[f'{year}년' for year in monthly_profit.index.levels[0]], loc='lower left', bbox_to_anchor=(1, 0.5))

plt.title('연도별 월 순수익')
plt.xlabel('시간(월)')
plt.ylabel('순수익(만원)')
plt.show()

monthly_profit

```

```python
# 결제 방법에 따른 분석 - 결제 방법과 결제금액의 연관성 분석  - 결제방법은 맨 앞의 한가지만 사용한 것으로 간주
shop_df['결제방법'].unique()
```

```python
# 신용카드(신 + 포, 신 + 적, 신 + 적 + 포, 신 + 웰컴)
# 가상계좌
# 후불(후 + 포, 후 + 적 + 포, 후 + 적)
# 무통장입금
# 포인트( + 포)
# 적립금(적 + 포)
# 웰컴마일
# 정기결제( 정 + 적. 정 + 적 + 포, 정 + 포)
# 현금간편결제(현+포, 현+적,현+적+포)
# - , Nan

# 금액과 저 위에 있는 결제방법
# 금액 방법 파이차트
```

```python
payment_mapping = {
    '신 + 포': '신용카드',
    '신 + 적': '신용카드',
    '신 + 적 + 포': '신용카드',
    '신 + 웰컴': '신용카드',
    '가상계좌': '가상계좌',
    '후 + 포': '후불',
    '후 + 적 + 포': '후불',
    '후 + 적': '후불',
    '무통장입금': '무통장입금',
    ' + 포': '포인트',
    '적 + 포': '적립금',
    '웰컴마일': '웰컴마일',
    '정 + 적': '정기결제',
    '정 + 적 + 포': '정기결제',
    '정 + 포': '정기결제',
    '현 + 포': '현금간편결제',
    '현 + 적': '현금간편결제',
    '현 + 적 + 포': '현금간편결제',
    '-': '기타',
    np.nan: '기타',
}

# 결제 방식을 새로운 카테고리로 변환
shop_df['결제카테고리'] = shop_df['결제방법'].map(payment_mapping)

# 카테고리별 거래 수 그래프
category_counts = shop_df['결제카테고리'].value_counts()
print(category_counts)

category_counts.plot(kind='bar', rot=90, color='lightcoral')
plt.title('결제 방식별 거래 수(건)')
plt.xlabel('결제 방식')
plt.ylabel('거래 수')
plt.show()
```

```python
# 주문한 달과 판매금액의 상관관계 분석
from sklearn.preprocessing import LabelEncoder

# 문자열 데이터를 숫자형으로 변환할 열 선택
categorical_columns = ['결제방법', '연도', '월', '업체명', '상품명' , '결제카테고리']  # 필요한 열을 추가하거나 변경하세요

# LabelEncoder 객체 생성
label_encoder = LabelEncoder()

# 선택한 열에 대해 숫자로 변환
for column in categorical_columns:
    shop_df[column] = label_encoder.fit_transform(shop_df[column].astype(str))

# 상관 계수 히트맵 다시 그리기
numeric_shop_df = shop_df.select_dtypes(include=[np.number])
correlation_matrix = numeric_shop_df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('상관 계수 히트맵')
plt.show()
```
